```python
import pandas as pd
import seaborn as sns
```

```python
l = open("../test.html").read()
```

```python
df = pd.read_html(l)[0]
```

```python
pop_list_1 = []
pop_list_2 = []
for i,j in zip(df["Population(2001)"].values, df["Population(2011)[3]"].values):
    try:
        pop_list_1.append(int(str(i)[0]))
    except:
        pass
    try:
        pop_list_2.append(int(str(j)[0]))
    except:
        pass
```

```python
sns.distplot(pop_list_1)
```

```python
sns.distplot(pop_list_2)
```
